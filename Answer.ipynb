```python
import matplotlib.pyplot as plt
a=[]
b=[]
with open('data.csv') as f:
    for line in f:
        x = float(line[0:line.index(',')])
        y = float(line[line.index(',')+1:len(line)])
        a.append(x)
        b.append(y)
for i in range(10):
    plt.clf()
    plt.plot(a,b)
    c=[]
    d=[]
    for i in range(1,len(a)):
        c.append((a[i-1]*3+a[i])/4)
        c.append((a[i-1]+a[i]*3)/4)
        d.append((b[i-1]*3+b[i])/4)
        d.append((b[i-1]+b[i]*3)/4)
    a=c
    b=d
```
