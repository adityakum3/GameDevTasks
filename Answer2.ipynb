```python
%reset -f             
from sympy import *    
from sympy.plotting import plot_parametric
from sympy.plotting import plot3d_parametric_line
import math
a=[]
b=[]
expr=[]
t = symbols('t')
with open('data.csv') as f:
    for line in f:
        x = float(line[0:line.index(',')])
        y = float(line[line.index(',')+1:len(line)])
        a.append(x)
        b.append(y)
for i in range(1,len(a)-1):
    x0=(a[i-1]+a[i])/2
    x1=a[i]
    x2=(a[i]+a[i+1])/2
    y0=(b[i-1]+b[i])/2
    y1=b[i]
    y2=(b[i]+b[i+1])/2
    x=(x0-2*x1+x2)*t*t+2*(x1-x0)*t+x0
    y=(y0-2*y1+y2)*t*t+2*(y1-y0)*t+y0
    expr.append((x, y, (t, 0, 1)))
plot_parametric(expr[0],expr[1],expr[2],expr[3],expr[4],expr[5],expr[6],expr[7],expr[8],expr[9],expr[10],expr[11],expr[12],expr[13],expr[14],expr[15],expr[16],expr[17],expr[18],expr[19],expr[20],expr[21],line_color='blue')
plt.savefig('foo.png')
#print(len(a))
#plot_parametric.show()
```
